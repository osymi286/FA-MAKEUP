```python
import skimage
from skimage import color
```

```python
#Lab에서 LCh값 변환
LCH_array=[]

for i in range(len(df.iloc[:])):
    lab = ([df['L*(D65)'][i],df['a*(D65)'][i],df['b*(D65)'][i]])
    lch = skimage.color.lab2lch(lab, channel_axis=-1)

    LCH_array.append(lch)
    i=i+1
```

```python
result=pd.DataFrame(LCH_array, columns=['L','C','h'])
result['h'] = result['h']*57.29578
result.to_csv('25024_Lab2LCh.csv')
```

```python
#Lab에서 RGB값 변환
rgb_array=[]

for i in range(len(df.iloc[:])):
    lab = ([df['L*(D65)'][i],df['a*(D65)'][i],df['b*(D65)'][i]])
    rgb = skimage.color.lab2rgb(lab, illuminant='D65', observer='2', channel_axis=-1)
    
    rgb_array.append(rgb)
    i=i+1
```

```python
result2=pd.DataFrame(rgb_array, columns=['r','g','b'])
result2.to_csv('25024_Lab2rgb.csv')
```

```python
#데이터프레임 병합
#https://kevinitcoding.tistory.com/entry/DataFrame-%EB%B3%91%ED%95%A9%ED%95%98%EA%B8%B0-concat-merge-join

LCH_array=[]
rgb_array=[]

for i in range(len(df.iloc[:])):
    lab = ([df['L*(D65)'][i],df['a*(D65)'][i],df['b*(D65)'][i]])
    lch = skimage.color.lab2lch(lab, channel_axis=-1)
    rgb = skimage.color.lab2rgb(lab, illuminant='D65', observer='2', channel_axis=-1)
       
    LCH_array.append(lch)
    rgb_array.append(rgb) 
    
    i=i+1

result=pd.DataFrame(LCH_array, columns=['L','C','h']) #LCh 변환값 column 3개만 있는 데이터프레임
result['h'] = result['h']*57.29578
result2=pd.DataFrame(rgb_array, columns=['r','g','b'])*255 #RGB 변환값 column 3개만 있는 데이터프레임

result_join=result.join(result2) #LCh 변환값과 RGB 변환값 병합 데이터프레임
result_total=df.join(result_join) #원본값과 LCh, RGB 변환값 병합 데이터프레임

result_total.to_csv('240923_맞춤형쿠션측색_total.csv', encoding='utf-8')
```

```python
tone1=result_total[result_total['Color'].str.contains('C3|C2|C1|N|W1|W2|W3')]['Color'].str[-2:]
tone1
```

```python
result_total.insert(15, 'TONE', tone1, allow_duplicates=False)
```

```python
result_total.dropna(axis=0, inplace=True)
```

```python
light=result_total['Color'].str[:-3]
light
```

```python
result_total.insert(16, 'LIGHT', light, allow_duplicates=False)
```

```python
#2025.02.17
# RGB 값을 HEX 코드로 변환하는 함수
def rgb_to_hex(rgb):
    # 실수 RGB 값을 0-255 범위의 정수로 변환
    r, g, b = [int(round(val)) for val in rgb]
    return "#{:02x}{:02x}{:02x}".format(r, g, b)
 
# numpy 배열을 순회하며 HEX 코드 리스트 생성
zerolip_in_hex_code = [rgb_to_hex(rgb) for rgb in df]

 
# HEX 코드 리스트를 DataFrame으로 변환
df_zerolip_in_hex_code = pd.DataFrame(zerolip_in_hex_code, columns=['HEX Code'])


#기존 df에 hex code 열로 추가
df_total = pd.concat([df, df_zerolip_in_hex_code], axis = 1)
df_total
```

```python
# RGB를 HEX로 변환하는 함수
def rgb_to_hex(rgb):
    # 실수 RGB 값을 0-255 범위의 정수로 변환
    r, g, b = [int(round(val)) for val in rgb]
    return "#{:02x}{:02x}{:02x}".format(r, g, b)

# 데이터 설정; 실제 DataFrame으로 대체하세요

# DataFrame에서 각 행의 RGB 값을 추출하여 hex 코드 리스트 생성
zerolip_in_hex_code = [rgb_to_hex((row['sR'], row['sG'], row['sB'])) for _, row in df.iterrows()]

# hex 코드 리스트를 DataFrame으로 변환
df_zerolip_in_hex_code = pd.DataFrame(zerolip_in_hex_code, columns=['hex code'])

# 기존 DataFrame에 새로운 hex code 열 추가
df_total = pd.concat([df, df_zerolip_in_hex_code], axis=1)

# 최종 결과 출력
print(df_total)

df_total.to_csv('250217hex후.csv', encoding='cp949')
```
